```python
# -*- coding: utf-8 -*-  
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#使用numpy加载文本文件中的数据
data=np.loadtxt("data/ex1data1.txt",delimiter=',')

#,创建绘图对象，也可以不创建绘图对象直接调用接下来的plot函数直接绘图，matplotlib会为我们自动创建一个绘图对象
#通过figsize参数可以指定绘图对象的宽度和高度，单位为英寸；dpi参数指定绘图对象的分辨率，即每英寸多少个像素，缺省值为80
plt.figure(figsize=(10,10),dpi=2048*2048)

#设置图表的标题
plt.title(u"机器学习课程样例数据")

#设置Y轴的范围
plt.ylim(-5,25)
#设置x轴的范围
plt.xlim(0,25)

#linewidth : 指定曲线的宽度
#label : 给所绘制的曲线一个名字，此名字在图示(legend)中显示。只要在字符串前后添加"$"符号，matplotlib就会使用其内嵌的latex引擎绘制的数学公式
#plt.plot(data[:,0],data[:,1],label=u"图示",color="red",linewidth=2)
#可以在第三个参数上指定颜色与格式，查看help(plt.plot)
plt.plot(data[:,0],data[:,1],"bD")

#设置Y轴的文字
plt.ylabel(u"Y值")
#设置x轴的文字
plt.xlabel(u"X值")
#显示图示
plt.legend()

#j显示绘图对象
plt.show()

#保存到文件
#plt.savefig('mylogsig.png',dpi=400,bbox_inches='tight')


```

```python
x=np.arange(0,5,0.5)
line, = plt.plot(x, x*x,'g')
#关闭对象的反锯齿效果
#line.set_antialiased(False)
#再加两条线并通过plt.setp函数配置多个Line2D对象的颜色和线宽属性
lines = plt.plot(x, np.sin(x), x, np.cos(x))
#设成红虚线，线宽2.0
plt.setp(lines, color="r",linestyle='--', linewidth=2.0)
plt.show()

plt.getp(lines[1])
## 只取'linestyle'一个属性
plt.getp(lines[1],'linestyle')


```

```python
#matplotlib的整个图表为一个Figure对象，此对象在调用plt.figure函数时返回，我们也可以通过plt.gcf函数获取当前的绘图对象
f=plt.gcf()

#Figure对象有一个axes属性，其值为AxesSubplot对象的列表，每个AxesSubplot对象代表图表中的一个子图，前面所绘制的图表只包含一个子图，当前子图也可以通过plt.gca获得
a=plt.gca()
#plt.setp(a,title="$机器学习课程样例数据")
alllines=plt.getp(a,"lines")
len(alllines)
```

```python
#一个绘图对象(figure)可以包含多个轴(axis)，在Matplotlib中用轴表示一个绘图区域，可以将其理解为子图。上面的第一个例子中，绘图对象只包括一个轴，因此只显示了一个轴(子图)。我们可以使用subplot函数快速绘制有多个轴的图表。subplot函数的调用形式如下：

#subplot(numRows, numCols, plotNum)
x=np.arange(0,5,0.5)
plt.subplot(221)
plt.plot(x,x*2,'r-')
plt.subplot(222)
plt.plot(x,x*3,'g--')
plt.subplot(223)
plt.plot(x,x*4,'yD-')
plt.subplot(224)
plt.plot(x,x*4,'g.-')
plt.show()
```
